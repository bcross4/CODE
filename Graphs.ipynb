```python
def func_run(DF: pd.DataFrame): 
    y_df = DF.iloc[:, 1]
    y_mean = np.mean(y_df)
    df_run = pd.DataFrame(columns=['Time', 'Value'])
    df_above_mean = pd.DataFrame(columns=['Time', 'Value'])
    df_below_mean = pd.DataFrame(columns=['Time', 'Value'])
    cnt_above = 0
    cnt_below = 0
    
    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0 and val['Value']>y_mean:
            df_above_mean = df_above_mean.append(val)
            cnt_above += 1 
            cnt_below = 0  
        elif idx==0 and val['Value']<=y_mean:
            df_below_mean = df_below_mean.append(val)
            cnt_above = 0 
            cnt_below += 1
        
        # second value
        if idx>0 and val['Value']>y_mean:
            df_above_mean = df_above_mean.append(val)
            cnt_above += 1 
            cnt_below = 0  
            # check if above threshold
            if cnt_above==8:
                df_run = df_run.append(df_above_mean[-cnt_above:]) 
            if cnt_above>8:
                ab = cnt_above - 1
                df_run = df_run[:-ab]
                df_run = df_run.append(df_above_mean[-cnt_above:]) 
        
        if idx>0 and val['Value']<=y_mean:
            df_below_mean = df_below_mean.append(val)
            cnt_above = 0 
            cnt_below += 1  
            # check if below threshold
            if cnt_below==8:
                df_run = df_run.append(df_below_mean[-cnt_below:])
            if cnt_below>8:
                bl = cnt_below - 1
                df_run = df_run[:-bl]
                df_run = df_run.append(df_below_mean[-cnt_below:])
    
    return df_run
```

```python
def func_trend(DF: pd.DataFrame):
    df_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_up_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_down_trend = pd.DataFrame(columns=['Time', 'Value'])
    df_previous_value = pd.DataFrame(columns=['Time', 'Value'])
    cnt_up = 0
    cnt_down = 0
    
    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0:
            df_previous_value = val 
        
        # next value compare previous value
        if idx>0 and val['Value']>=df_previous_value['Value']:
            df_previous_value = val
            df_up_trend = df_up_trend.append(val)
            cnt_up += 1 
            cnt_down = 0 
            # check if above threshold
            if cnt_up==5: 
                df_trend = df_trend.append(DF.iloc[idx-5])
                df_trend = df_trend.append(df_up_trend[-cnt_up:]) 
            if cnt_up>5:
                ab = cnt_up - 1
                df_trend = df_trend[:-ab]
                df_trend = df_trend.append(df_up_trend[-cnt_up:])
    
        if idx>0 and val['Value']<df_previous_value['Value']:
            df_previous_value = val
            df_down_trend = df_down_trend.append(val)
            cnt_up = 0
            cnt_down += 1 
            # check if above threshold
            if cnt_down==5:
                df_trend = df_trend.append(DF.iloc[idx-5])
                df_trend = df_trend.append(df_down_trend[-cnt_down:]) 
            if cnt_down>5:
                bl = cnt_down - 1
                df_trend = df_trend[:-bl]
                df_trend = df_trend.append(df_down_trend[-cnt_down:]) 
    
    return df_trend
```

```python
def func_hug(DF: pd.DataFrame):
    df_zone = pd.DataFrame(columns=['Time', 'Value'])
    df_zone_c = pd.DataFrame(columns=['Time', 'Value'])
    df_not_zone_c = pd.DataFrame(columns=['Time', 'Value'])
    cnt_zone_c = 0 
    
    y = DF['Value']
    y_mean = np.mean(y)
    std=DF['Value'].std()
    std_ucl_one=y_mean+(std*1) 
    std_lcl_one=y_mean-(std*1) 
    
    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==0:
            df_previous_value = val 
        
        # next value compare previous value
        if idx>0 and std_lcl_one <= df_previous_value['Value'] <= std_ucl_one and std_lcl_one <= val['Value'] <= std_ucl_one:
            df_previous_value = val
            df_zone_c = df_zone_c.append(val)
            cnt_zone_c += 1  
            # check if above threshold
            if cnt_zone_c==15: 
                df_zone = df_zone.append(DF.iloc[idx-15])
                df_zone = df_zone.append(df_zone_c[-cnt_zone_c:]) 
            if cnt_zone_c>15:
                ab = cnt_zone_c - 1
                df_zone = df_zone[:-ab]
                df_zone = df_zone.append(df_zone_c[-cnt_zone_c:])
        else:
            df_previous_value = val
            cnt_zone_c = 0  
    
    return df_zone
```

```python
def func_4_5(DF: pd.DataFrame):
    df_4_5 = pd.DataFrame(columns=['Time', 'Value'])
    cnt_over_zone_c = 0 
    cnt_under_zone_c = 0 
    
    y = DF['Value']
    y_mean = np.mean(y)
    std=DF['Value'].std()
    std_ucl_one=y_mean+(std*1) 
    std_lcl_one=y_mean-(std*1) 
    
    for idx, val in DF.iterrows():
        # is this is the first value in the dataframe?
        if idx==4:
            cnt_over_zone_c  = len([i for i in DF[:5]['Value'] if i > std_ucl_one])
            cnt_under_zone_c = len([i for i in DF[:5]['Value'] if i < std_lcl_one])
            if cnt_over_zone_c>=4 or cnt_under_zone_c>=4:
                df_4_5 = df_4_5.append(DF[:5]) 
        
        if idx>4:
            cnt_over_zone_c  = len([i for i in DF[idx-4:idx+1]['Value'] if i > std_ucl_one])
            cnt_under_zone_c = len([i for i in DF[idx-4:idx+1]['Value'] if i < std_lcl_one])
            if cnt_over_zone_c>=4 or cnt_under_zone_c>=4:
                df_4_5 = df_4_5[:-4]
                df_4_5 = df_4_5.append(DF[idx-4:idx+1])
        
    return df_4_5
```

```python
def func_control_chart(DF: pd.DataFrame):
    y = DF['Value']
    y_mean = np.mean(y)
    
    # OUTLIERS
    q1=DF['Value'].quantile(0.25) 
    q3=DF['Value'].quantile(0.75)
    iqr=q3-q1 
    lcl = q1-(1.5*iqr)
    ucl = q3+(1.5*iqr)
    btw_cl = DF['Value'].between(lcl,ucl)
    df_outliers = DF[~btw_cl]
    
    # SIGMA ZONES
    std=DF['Value'].std()
    std_ucl_one=y_mean+(std*1)
    std_ucl_two=y_mean+(std*2)
    std_ucl_three=y_mean+(std*3)
    std_lcl_one=y_mean-(std*1)
    std_lcl_two=y_mean-(std*2)
    std_lcl_three=y_mean-(std*3)
    
    # PLOT 
    fig, ax = plt.subplots(1, figsize=(16, 6)) 
    ax.set_title('Control Chart', fontweight='bold', fontsize=26)
    ax.set_xlabel('Time', fontweight='bold')
    ax.set_ylabel('Values', fontweight='bold')
    ax.set_xlim(np.min(DF['Time']), np.max(DF['Time']))
    ax.set_ylim(np.min(DF['Value']-10), np.max(DF['Value']+10))
    plt.axhspan(std_lcl_one, std_ucl_one, facecolor='g', alpha=0.5) 
    plt.axhspan(std_lcl_two, std_ucl_two, facecolor='g', alpha=0.3) 
    plt.axhspan(std_lcl_three, std_ucl_three, facecolor='g', alpha=0.2) 
    plt.axhline(y=y_mean, color='k')
    ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
    ax.xaxis.set_major_formatter(DateFormatter("%Y-%m-%d"))
    
    # all 
    plt.plot(DF['Time'], DF['Value'], linestyle='--',linewidth=0.6, marker='.', markersize=5,color='b', label='all data')
    # outliers
    plt.plot(df_outliers['Time'], df_outliers['Value'], linestyle='none', marker='^', markersize=8, color='r', label='outliers')
    # run 
    plt.plot(df_run['Time'], df_run['Value'], linestyle='none', marker='o', markersize=7, color='r', label='run')
    # trend 
    plt.plot(df_trend['Time'], df_trend['Value'], linestyle='none', marker='>', markersize=5, color='r', label='trend')
    # hugging 
    plt.plot(df_hug['Time'], df_hug['Value'], linestyle='none', marker='P', markersize=7, color='r', label='hugging')
    # 4 out of 5 are above/below 1 std
    plt.plot(df_4_5['Time'], df_4_5['Value'], linestyle='none', marker='D', markersize=5, color='r', label='4_5')
    
    ax.legend(loc=2)
    fig.autofmt_xdate() 
    plt.tight_layout()
    
    return plt.show()
```

# Waterfall

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'category':['Sales','Service','Expenses','Taxes','Interest'],
                   'num':[100,10,-20,-30,60]})
df
```

```python
def waterfall(df, x, y):
    
    plt.rcParams["figure.figsize"] = (16,8)  
    
    # calculate running totals
    df['tot'] = df[y].cumsum()
    df['tot1']=df['tot'].shift(1).fillna(0)


    # lower and upper points for the bar charts
    lower = df[['tot','tot1']].min(axis=1)
    upper = df[['tot','tot1']].max(axis=1)


    # mid-point for label position
    mid = (lower + upper)/2

    # positive number shows green, negative number shows red
    df.loc[df[y] >= 0, 'color'] = 'green'
    df.loc[df[y] < 0, 'color'] = 'red'

    # calculate connection points
    connect= df['tot1'].repeat(3).shift(-1)
    connect[1::3] = np.nan

    fig,ax = plt.subplots()

    # plot first bar with colors
    bars = ax.bar(x=df[x],height=upper, color =df['color'])

    # plot second bar - invisible
    plt.bar(x=df[x], height=lower,color='white')

    # plot connectors
    plt.plot(connect.index,connect.values, 'k' )

    # plot bar labels
    for i, v in enumerate(upper):
        plt.text(i-.15, mid[i], f"{df[y][i]:,.0f}")
```

```python
waterfall(df,'category','num')
```
